```python
import numpy as np

# Definimos matrices
A = np.array([[2, 3],
              [1, 4]])

B = np.array([[1, 2],
              [3, 1]])

print("Matriz A:\n", A)
print("Matriz B:\n", B)

# 1. Determinante bsico
detA = np.linalg.det(A)
print("\n1. Determinante de A:", detA)

# 2. Si det(A) != 0, la matriz es invertible
try:
    invA = np.linalg.inv(A)
    print("2. Matriz A es invertible, inversa:\n", invA)
except np.linalg.LinAlgError:
    print("2. Matriz A no es invertible (det=0)")

# 3. Propiedad multiplicacin: det(AB) = det(A)*det(B)
detAB = np.linalg.det(A @ B)
detA_detB = detA * np.linalg.det(B)
print("\n3. Propiedad multiplicacin:")
print("det(AB):", detAB, "  vs  det(A)*det(B):", detA_detB)

# 4. Propiedad de la transpuesta: det(A^T) = det(A)
detAT = np.linalg.det(A.T)
print("\n4. Propiedad transpuesta:")
print("det(A^T):", detAT, "  vs  det(A):", detA)

# 5. Relacin con eigenvalores: det(A) = producto de eigenvalores
eigvals = np.linalg.eigvals(A)
prod_eigvals = np.prod(eigvals)
print("\n5. Relacin con eigenvalores:")
print("det(A):", detA, "  vs  producto de eigenvalores:", prod_eigvals)
```

```python
import numpy as np

# Definimos matrices cuadradas
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

c = 3  # escalar

print("Matriz A:\n", A)
print("Matriz B:\n", B)

# 1. Linealidad: Tr(A + B) = Tr(A) + Tr(B), y Tr(cA) = c * Tr(A)
tr_lineal = np.trace(A + B)
tr_sum = np.trace(A) + np.trace(B)
tr_escalar = np.trace(c * A)
tr_escalar_prop = c * np.trace(A)

print("\n1. Linealidad:")
print("Tr(A + B):", tr_lineal, "  vs  Tr(A)+Tr(B):", tr_sum)
print("Tr(cA):", tr_escalar, "  vs  c*Tr(A):", tr_escalar_prop)

# 2. Invarianza por transpuesta: Tr(A^T) = Tr(A)
tr_transpuesta = np.trace(A.T)
print("\n2. Invarianza por transpuesta:")
print("Tr(A^T):", tr_transpuesta, "  vs  Tr(A):", np.trace(A))

# 3. Propiedad cclica: Tr(AB) = Tr(BA)
tr_AB = np.trace(A @ B)
tr_BA = np.trace(B @ A)
print("\n3. Propiedad cclica:")
print("Tr(AB):", tr_AB, "  vs  Tr(BA):", tr_BA)

# 4. Relacin con autovalores: Tr(A) = suma de eigenvalores
eigenvals = np.linalg.eigvals(A)
print("\n4. Relacin con autovalores:")
print("Tr(A):", np.trace(A), "  vs  Suma eigenvalores:", np.sum(eigenvals))
```

```python
import numpy as np

# Definimos matriz cuadrada
X = np.array([[2.0, 1.0],
              [1.0, 3.0]])

A = np.array([[1.0, 2.0],
              [0.0, -1.0]])

B = np.array([[2.0, 0.0],
              [1.0, 1.0]])

print("Matriz X:\n", X)
print("Matriz A:\n", A)
print("Matriz B:\n", B)

# 1. Gradiente de Tr(AX)
grad1 = A.T
print("\n1. Gradiente de Tr(AX):\n", grad1)

# 2. Gradiente de Tr(AXB)
grad2 = A.T @ B.T
print("\n2. Gradiente de Tr(AXB):\n", grad2)

# 3. Gradiente de ||X||_F^2 = Tr(X^T X)
grad3 = 2 * X
print("\n3. Gradiente de ||X||_F^2:\n", grad3)

# 4. Gradiente de det(X)
detX = np.linalg.det(X)
grad4 = detX * np.linalg.inv(X).T
print("\n4. Gradiente de det(X):\n", grad4)

# 5. Gradiente de log(det(X))
grad5 = np.linalg.inv(X).T
print("\n5. Gradiente de log(det(X)):\n", grad5)

# 6. Gradiente de ||X||_2 (norma Eucldea de todos los elementos)
norm2 = np.linalg.norm(X)  # Frobenius norm
grad6 = X / norm2
print("\n6. Gradiente de ||X||_2:\n", grad6)
```
